```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/bankruptcy_data.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
print("Total Missing Values:", df.isnull().sum().sum())
```

```python
print(df['Bankrupt?'].value_counts())
print(df['Bankrupt?'].value_counts(normalize=True) * 100)
```

```python
df['Bankrupt?'].value_counts().plot(kind='bar')
plt.title("Bankruptcy Class Distribution")
plt.show()
```

```python
df.describe().T
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
print("New Shape:", df.shape)
```

```python
X = df.drop(columns=['Bankrupt?'])
y = df['Bankrupt?']

print("Feature Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

```python
import numpy as np

corr_matrix = X_train.corr().abs()

print("Correlation matrix shape:", corr_matrix.shape)
```

```python
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)
```

```python
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]

print("Number of highly correlated features:", len(to_drop))
print("Features to drop:")
print(to_drop[:10])  # show first 10
```

```python
X_train = X_train.drop(columns=to_drop)
X_test = X_test.drop(columns=to_drop)

print("New feature count after correlation removal:", X_train.shape[1])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling completed.")
print("Scaled Train Shape:", X_train_scaled.shape)
print("Scaled Test Shape:", X_test_scaled.shape)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_balanced, y_train_balanced = smote.fit_resample(
    X_train_scaled,
    y_train
)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", pd.Series(y_train_balanced).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train_balanced, y_train_balanced)

print("Logistic Regression model trained.")
```

```python
y_pred = log_model.predict(X_test_scaled)
y_prob = log_model.predict_proba(X_test_scaled)[:, 1]
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_prob))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train_balanced, y_train_balanced)

print("Random Forest trained.")
```

```python
y_pred_rf = rf_model.predict(X_test_scaled)
y_prob_rf = rf_model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_prob_rf))
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

xgb_model.fit(X_train_balanced, y_train_balanced)

print("XGBoost trained.")
```

```python
y_pred_xgb = xgb_model.predict(X_test_scaled)
y_prob_xgb = xgb_model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_prob_xgb))
```

```python
threshold = 0.30

y_pred_custom = (y_prob_xgb > threshold).astype(int)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_custom))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_custom))
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_prob_xgb)

plt.figure(figsize=(8,6))
plt.plot(thresholds, recall[:-1], label="Recall")
plt.plot(thresholds, precision[:-1], label="Precision")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.title("Precision-Recall vs Threshold")
plt.show()
```

```python
def calculate_risk_score(probability):
    return float(round(probability * 100, 2))
# Example
risk_scores = [calculate_risk_score(float(p)) for p in y_prob_xgb[:10]]
print("Sample Risk Scores:", risk_scores)
```

```python
def risk_category(score):
    if score < 30:
        return "Low Risk"
    elif score < 60:
        return "Medium Risk"
    else:
        return "High Risk"

# Example
for s in risk_scores[:5]:
    print(s, risk_category(s))
```

```python
def predict_company_risk(input_data):
    # Scale input
    input_scaled = scaler.transform(input_data)
    
    # Predict probability
    prob = float(xgb_model.predict_proba(input_scaled)[:, 1][0])
    
    # Convert to score
    score = round(prob * 100, 2)
    
    # Assign category
    if score < 30:
        category = "Low Risk"
    elif score < 60:
        category = "Medium Risk"
    else:
        category = "High Risk"
    
    return {
        "Probability": round(prob, 4),
        "Risk Score": score,
        "Risk Category": category
    }
```

```python
sample = X_test.iloc[[0]]
result = predict_company_risk(sample)

print(result)
```
